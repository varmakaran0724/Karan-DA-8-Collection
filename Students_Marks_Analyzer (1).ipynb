Creating a Student Marks Analyzer for assessing Grades

```python
import numpy as np

def subject_marks(average_marks):
#  Classifying marks and grade based on the scores of the student

    if average_marks >= 85:
        return "A (Outstanding)"
    elif average_marks >= 70:
        return "B (Good)"
    elif average_marks >= 50:
        return "C (Average)"
    elif average_marks >= 40:  # below 40 is fail
        return "D (Pass)"
    else:
        return "Fail (Needs Academic Review)"

print("=== MARKS ANALYZER ===")

# Creation of a list to put in subjects that marks need be taken
subject_type = ["Maths", "English", "Science", "Physical Education"]
marks_list = []

print("Please enter subject marks (out of 100):\n")

# Collect subject marks
for subject in subject_type:
    score = float(input(f"Enter score obtained in {subject}: "))
    marks_list.append(score)

# Converting structured Python list data to a NumPy Array
marks_array = np.array(marks_list)

# Mathematical array operations using NumPy methods
total_marks = np.sum(marks_array)
average_marks = np.mean(marks_array)
Max_mark = np.max(marks_array)
Min_mark = np.min(marks_array)

# Condition tier classification mapping
assigned_grade = subject_marks(average_marks)

# Finding the indices of max and min values using NumPy
highest_subject = subject_type[int(np.where(marks_array == Max_mark)[0][0])]
lowest_subject = subject_type[int(np.where(marks_array == Min_mark)[0][0])]

# Printing final comprehensive card report
print("\n_________________________________________")
print("         STUDENT REPORT CARD - 2026    ")
print("_________________________________________")
for i in range(len(subject_type)):
    print(f"  {subject_type[i]}: {marks_array[i]:.2f} / 100.00")
print("_________________________________________")
print(f" Total Marks:  {total_marks:.2f} / 400.00")                    #for total marks
print(f" Calculated Percentage: {average_marks:.2f}%")                 #for percentage
print(f" Strongest Area:          {highest_subject} ({Max_mark:.2f})") #for highest score among the subjects
print(f" Needs Work Area:         {lowest_subject} ({Min_mark:.2f})")  #for lowest score among the subjects
print(f" Assigned Final Grade:    {assigned_grade}")
print("_________________________________________")
```
